```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv = pd.read_csv("E:\\Data Analyst (Project)\\Construction excel for jupyter.csv")
csv
```

```python
csv['Budget (INR)'] = csv['Budget (INR)'].str.replace('?', '').str.replace(',', '').astype(float)
csv['Actual Cost (INR)'] = csv['Actual Cost (INR)'].str.replace('?', '').str.replace(',', '').astype(float)
```

In this step, we are cleaning the `Budget (INR)` and `Actual Cost (INR)` columns in our dataset. These columns originally contain special characters such as commas (`,`) used as thousand separators and question marks (`?`) that may represent missing or invalid values. 

```python
csv['Start Date'] = pd.to_datetime(csv['Start Date'], dayfirst=True)
csv['Planned End Date'] = pd.to_datetime(csv['Planned End Date'], dayfirst=True)
csv['Actual End Date'] = pd.to_datetime(csv['Actual End Date'], dayfirst=True)
```

In this step, we are converting the columns to proper datetime format:

We use the `pd.to_datetime()` function with the parameter `dayfirst=True` because the date format in our dataset appears to follow the **DD-MM-YYYY** format (common in India and Europe). This ensures accurate parsing of dates.

```python
sns.barplot(data = csv, x = "Location", y = "Site Incidents",edgecolor ="black",capsize=0.3,color="grey")
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

###  Bar Plot: Average Site Incidents by Location

This bar chart visualizes the **Average No. of Site Incidents** for each project location.

 What this chart tells us:
- Each bar represents the mean number of site incidents at each location.

- The black lines (error bars) show the confidence intervalgiving us an idea of how much the incident values vary.

- Locations like Chennai and Hyderabad have the highest average number of incidents.

- Locations like Delhi and Bangalore show lower average incidents.

```python
plt.figure(figsize=(8,5))
plt.hist(x="Actual Duration (days)",data=csv,edgecolor="black")
plt.title("Distribution of Actual Day")
plt.xlabel("Days")
plt.show()
```

###  Histogram: Distribution of Actual Project Duration (Days)

This histogram shows how the**Actual Duration**  (in days) of construction projects is distributed across the dataset

 What this chart tells us:
- The x-axis (Days) shows the range of project durations.

- The y-axis shows how many projects fall into each duration range.

- Most projects tend to have actual durations between 250 to 450 days.

- A few projects extended close to 600 days, which could indicate delays or larger-scale projects.


```python
sns.histplot(x="Average Daily Labor",data=csv,edgecolor="black",color="darkblue")
plt.show()
```

###  Histogram: Distribution of Average Daily Labor

This histogram visualizes the **Distribution of Average Daily labor count** across construction projects.

 What this chart tells us:
- The x-axis shows the average number of laborers used per day on each project.

- The y-axis shows the number of projects that fall within each labor range.

- Most projects seem to have between 40 to 70 daily workers.

- A smaller number of projects used over 100 laborers daily, suggesting larger-scale or faster-paced projects.

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Location', data=csv,edgecolor="black")
plt.xlabel("location of project")
plt.ylabel("count of project")
plt.title("Count of projects in each Location")
plt.tight_layout()
plt.show()
```

###  Bar Chart: Count of Projects in Each Location

This bar chart Displays the **Number of Construction Projects** across different cities in the dataset.

 What this chart tells us:
- Each bar shows how many projects were conducted in a specific location.

- Hyderabad has the highest number of projects, followed by Kolkata, Chennai, and Bangalore.

- Pune and Delhi have the fewest projects in this dataset.

```python
sns.boxplot(x='Location', y='Average Daily Labor', data=csv)
plt.tight_layout()
plt.title("Comparing Labor distribution across Locations")
plt.show()
```

###  Boxplot: Comparing Average Daily Labor Across Locations

This Boxplot visualizes the **Distribution of Average Daily labor** across different project locations, highlighting variations and outliers.

 What this chart tells us:
- Each box represents the interquartile range (IQR) (25th to 75th percentile) of daily labor used in each city.

- The line inside each box shows the median.

- Whiskers extend to show the range within 1.5IQR.

- Dots (like in Delhi) are outliers, indicating unusual labor counts.

- Chennai shows a wide spread, meaning large variation in labor use across projects.

- Delhi has a small IQR, suggesting very consistent labor numbers.
